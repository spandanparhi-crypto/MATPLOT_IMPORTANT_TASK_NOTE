```python
# import lib matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
# data

X = [1,2,3,4,5]

Y = [10,11,12,13,14]


# make the simple plot 

plt.plot(X,Y)
plt.show()

```

```python
# pylot with label and design 


plt.plot(X,
         Y,
         color= "green",
         linestyle = "--",
         linewidth = 2,
         marker = "o",
         markersize = 10,
         markerfacecolor = "red",
         markeredgecolor = "black",
         markeredgewidth = 2,
         alpha  = 1.0,
         
        )

plt.title("Plot the Data ") # title of plot
plt.xlabel("X Axis")        # lable of x axis
plt.ylabel("Y Axis")        # label of y axis
plt.grid()
plt.show()
```

```python
# Pyplot Common Commands 

# make the data 

x = np.linspace(0,10,100)

y = np.sin(x)

# make common commands

# step1:  make figure size

plt.figure(figsize=(5,5))

# step 2: make the plot
plt.plot( x,y  ,      
         color= "green",
         linestyle = "--",
         linewidth = 2,
         marker = "o",
         markersize = 10,
         markerfacecolor = "red",
         markeredgecolor = "black",
         markeredgewidth = 2,
         alpha  = 1.0,)

# step 3: title of graph 
plt.title("My first Graph")

# step 4: X axis label
plt.xlabel("X axis")

# Step 5: Y axis label
plt.ylabel("Y axis")

# Step 6: set X axis range
plt.xlim(0,10)

# Step 7: set y axis range
plt.ylim(-1.5,1.5)

# step 8: set the x axis ticks 
ranges=np.arange(0,11,2)

plt.xticks(ranges)

# step 9: set the y axis ticks
plt.yticks([-1,0,1])

# Step 10 : set the grid of graph
plt.grid(True, alpha=.7)


# Step 11 : Set the legend
plt.legend(["Sale"])

# Step 12 : set the tight_layout to graph : auto fix spacing
plt.tight_layout() 

# Show the plot 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Make the data
X = np.linspace(0, 10, 100)

# Create figure with multiple subplots
fig, ax = plt.subplots(2, 2, figsize=(5,5))

# Figure title
fig.suptitle(
    "Figure with 4 Subplots",
    fontsize=16,
    fontweight="bold"
)

# First subplot
ax[0, 0].plot(X, np.sin(X), color="blue", marker="o")
ax[0, 0].set_title("Sin Plot")

# Add text annotation 
ax[0, 0].text(9, -1.0, "Sin plot",fontsize=5, color='red', ha='center')

# Add horizontal line 
ax[0, 0].axhline(y=0.5, color='black', 
           linewidth=1, linestyle='--') 



# Second subplot
ax[0, 1].plot(X, np.cos(X), color="green", marker="o")
ax[0, 1].set_title("Cos Plot")

# Third subplot
ax[1, 0].plot(X, np.tan(X), color="red", marker="o")
ax[1, 0].set_title("Tan Plot")

# Fourth subplot
ax[1, 1].plot(X, X**2, color="purple", marker="o")
ax[1, 1].set_title("Squared Plot")

# Adjust spacing
plt.tight_layout()

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Make the data
X = np.linspace(0, 10, 100)

# Create figure with multiple subplots
fig, ax = plt.subplots(1, figsize=(5,5))

# Figure title
fig.suptitle(
    "Figure with 4 Subplots",
    fontsize=16,
    fontweight="bold"
)

# First subplot
ax.plot(X, np.sin(X), color="blue", marker="o")
ax.set_title("Sin Plot")

# Add text annotation 
ax.text(9, -1.0, "Sin plot",fontsize=5, color='red', ha='center')

# Add horizontal line 
ax.axhline(y=0.5, color='black', 
           linewidth=1, linestyle='--') 


# Setting Limits and Tick Labels 
ax.set_xlim(1,10)        # show only x from 5 to 15 
ax.set_ylim(-0.5, 0.5)    # show only y from -0.5 to 0.5 
ax.set_title("Custom Limits (Zoomed In!)") 
ax.grid(True) 

# Setting Tick Positions

ax.set_xticklabels( 
    ["A","B","C","D","E","F","g","H","I","J"], 
    rotation=45,           # rotate labels 45 degrees 
    fontsize=9 
)  
ax.set_yticks([-1, -0.5, 0, 0.5, 1])


plt.show()
```
