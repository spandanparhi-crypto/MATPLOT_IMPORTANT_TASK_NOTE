```python
import numpy as np
import matplotlib.pyplot as plt

# Create the data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create figure
plt.figure(figsize=(6, 5))

# Plot the sine wave
plt.plot(
    x,
    y,
    color="green",
    linewidth=2,
    linestyle="--",
    label="sin(x)"
)

# Add title and labels
plt.title("Sine Wave")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Set axis limits
plt.xlim(0, 10)
plt.ylim(-1.5, 1.5)

# Add grid
plt.grid(True)

# Add legend
plt.legend()

# Display the graph
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

# Create figure
plt.figure(figsize=(12, 6))

# Plot the graph
plt.plot(
    x,
    y,
    color="blue",
    linewidth=2,
    marker="o",
    markersize=8,
    label="Sales Data"
)

# Title
plt.title("Sales Data Graph")

# Labels
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Grid
plt.grid(True)

# Legend
plt.legend()

# Adjust layout
plt.tight_layout()

# Show graph
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for Line A
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 30, 40, 50]

# Data for Line B
y2 = [5, 15, 10, 25, 20]

# Create Figure
plt.figure(figsize=(8, 5))

# Plot Line A
plt.plot(
    x,
    y1,
    color="blue",
    linewidth=2,
    marker="o",
    label="Line A"
)

# Plot Line B
plt.plot(
    x,
    y2,
    color="red",
    linewidth=2,
    marker="s",
    linestyle="--",
    label="Line B"
)

# Title
plt.title("Two Lines in One Graph")

# Labels
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Grid
plt.grid(True)

# Legend
plt.legend()

# Show Graph
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [5, 10, 15, 20, 25]

# Create Figure
plt.figure(figsize=(10, 6))

# Plot the graph
plt.plot(
    x,
    y,
    color="purple",
    linewidth=2,
    marker="o",
    markersize=8,
    markerfacecolor="yellow",
    markeredgecolor="black",
    label="Student Data"
)

# Title
plt.title("Student Plot", fontsize=16, fontweight="bold")

# Axis Labels
plt.xlabel("X Axis", fontsize=12)
plt.ylabel("Y Axis", fontsize=12)

# Grid
plt.grid(True, alpha=0.3)

# Legend
plt.legend()

# Save the figure
plt.savefig("student_plot.png", dpi=150)

# Adjust layout
plt.tight_layout()

# Display the graph
plt.show()
```
