```python
# . Create Multiple Subplots

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,100)

fig, ax = plt.subplots(2,2,figsize=(8,6))

ax[0,0].plot(x,np.sin(x))
ax[0,0].set_title("Sin")

ax[0,1].plot(x,np.cos(x))
ax[0,1].set_title("Cos")

ax[1,0].plot(x,np.tan(x))
ax[1,0].set_title("Tan")

ax[1,1].plot(x,x**2)
ax[1,1].set_title("Square")

plt.tight_layout()
plt.show()
```

```python
# Shared Axis

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,100)

fig, ax = plt.subplots(2, 1, sharex=True)

ax[0].plot(x, np.sin(x))
ax[0].set_title("Sine")

ax[1].plot(x, np.cos(x))
ax[1].set_title("Cosine")

plt.tight_layout()

plt.show()
```

```python
# Annotate a Point

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,200)

y = np.sin(x)

plt.plot(x,y)

max_index = np.argmax(y)

plt.annotate(
    "Maximum",
    xy=(x[max_index],y[max_index]),
    xytext=(6,1.2),
    arrowprops=dict(facecolor='red')
)

plt.grid(True)

plt.show()
```

```python
# Compare Two Datasets

import matplotlib.pyplot as plt

months=["Jan","Feb","Mar","Apr","May","Jun"]

companyA=[120,140,160,180,200,220]
companyB=[100,130,150,170,210,230]

plt.plot(
    months,
    companyA,
    marker="o",
    linewidth=2,
    label="Company A"
)

plt.plot(
    months,
    companyB,
    marker="s",
    linewidth=2,
    linestyle="--",
    label="Company B"
)

plt.title("Monthly Sales Comparison")

plt.xlabel("Months")
plt.ylabel("Sales")

plt.grid(True)

plt.legend()

plt.show()
```

```python
# Use subplot()

import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,10,100)

plt.figure(figsize=(8,6))

plt.subplot(2,2,1)
plt.plot(x,np.sin(x))
plt.title("Sin")

plt.subplot(2,2,2)
plt.plot(x,np.cos(x))
plt.title("Cos")

plt.subplot(2,2,3)
plt.plot(x,np.tan(x))
plt.title("Tan")

plt.subplot(2,2,4)
plt.plot(x,x**2)
plt.title("Square")

plt.tight_layout()

plt.show()
```

```python
# Complete Dashboard

import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,10,100)

fig,ax=plt.subplots(2,2,figsize=(12,8))

fig.suptitle("Matplotlib Dashboard",fontsize=18)

# Sin Graph
ax[0,0].plot(x,np.sin(x),color="blue",label="Sin")
ax[0,0].axhline(0,color="black",linestyle="--")
ax[0,0].text(7,0.8,"Sine Wave",color="red")
ax[0,0].set_title("Sin Graph")
ax[0,0].set_xlabel("X")
ax[0,0].set_ylabel("Sin(X)")
ax[0,0].legend()
ax[0,0].grid(True)

# Cos Graph
ax[0,1].plot(x,np.cos(x),color="green",label="Cos")
ax[0,1].axhline(0,color="black",linestyle="--")
ax[0,1].text(7,0.8,"Cos Wave",color="red")
ax[0,1].set_title("Cos Graph")
ax[0,1].set_xlabel("X")
ax[0,1].set_ylabel("Cos(X)")
ax[0,1].legend()
ax[0,1].grid(True)

# Scatter Plot
ax[1,0].scatter(
    np.random.rand(50),
    np.random.rand(50),
    color="purple",
    label="Points"
)
ax[1,0].set_title("Scatter Plot")
ax[1,0].set_xlabel("X")
ax[1,0].set_ylabel("Y")
ax[1,0].legend()
ax[1,0].grid(True)

# Histogram
data=np.random.randn(1000)

ax[1,1].hist(
    data,
    bins=20,
    color="orange",
    edgecolor="black",
    label="Frequency"
)

ax[1,1].set_title("Histogram")
ax[1,1].set_xlabel("Values")
ax[1,1].set_ylabel("Frequency")
ax[1,1].legend()
ax[1,1].grid(True)

plt.tight_layout(rect=[0,0,1,0.95])

plt.show()
```
