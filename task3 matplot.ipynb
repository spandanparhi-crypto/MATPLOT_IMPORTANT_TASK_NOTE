```python
# 1. How to Make One Chart (Single Plot)

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.plot(x, y)

plt.show()
```

```python
# 2. How to Make One Single Axes

import matplotlib.pyplot as plt

fig, ax = plt.subplots()

x = [1, 2, 3, 4]
y = [5, 10, 15, 20]

ax.plot(x, y)

plt.show()
```

```python
# 3. How to Make Multiple Plots (Subplots)

import matplotlib.pyplot as plt

x = [1,2,3,4]
y1 = [10,20,30,40]
y2 = [40,30,20,10]

plt.subplot(1,2,1)
plt.plot(x, y1)
plt.title("First Plot")

plt.subplot(1,2,2)
plt.plot(x, y2)
plt.title("Second Plot")

plt.show()
```

```python
# 4. How to Make Multiple Plots in a Single Axes

import matplotlib.pyplot as plt

x = [1,2,3,4]

y1 = [10,20,30,40]
y2 = [40,30,20,10]

plt.plot(x, y1)
plt.plot(x, y2)

plt.show()
```

```python
# 5. How to Give Labels, X Label, Y Label, Legend and Title

import matplotlib.pyplot as plt

x = [1,2,3,4,5]

sales = [20,35,40,45,60]

plt.plot(x, sales, label="Sales")

plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Monthly Sales")
plt.legend()

plt.show()
```

```python
# 6. Complete Example (Recommended)

import matplotlib.pyplot as plt

months = ['Jan','Feb','Mar','Apr','May']
sales = [100,120,140,130,160]
profit = [20,30,35,40,50]

plt.figure(figsize=(8,5))

plt.plot(months, sales, marker='o', label='Sales')
plt.plot(months, profit, marker='s', label='Profit')

plt.xlabel("Months")
plt.ylabel("Amount")
plt.title("Monthly Sales and Profit")

plt.legend()

plt.grid(True)

plt.show()
```

```python
# 7. Using fig and ax (Professional Method

import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8,5))

x = [1,2,3,4,5]
y = [10,15,12,20,25]

ax.plot(x, y, marker='o', label="Data")

ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_title("Simple Line Graph")

ax.legend()
ax.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [120, 140, 135, 160, 180, 200]
profit = [30, 40, 35, 50, 60, 70]

# Create Figure and Axes
fig, ax = plt.subplots(figsize=(8, 5))

# Plot Multiple Lines on the Same Axes
ax.plot(months, sales, marker='o', linestyle='-', color='blue', label='Sales')
ax.plot(months, profit, marker='s', linestyle='--', color='red', label='Profit')

# X-axis Label
ax.set_xlabel("Months", fontsize=12)

# Y-axis Label
ax.set_ylabel("Amount", fontsize=12)

# Title
ax.set_title("Monthly Sales and Profit Report", fontsize=14)

# Legend
ax.legend()

# Grid
ax.grid(True)

# Display Graph
plt.show()
```

```python
# Complete Example with Multiple Subplots

import matplotlib.pyplot as plt

# Data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [100, 120, 140, 130, 160]
profit = [20, 30, 35, 40, 50]

# Create Figure
plt.figure(figsize=(10, 4))

# First Subplot
plt.subplot(1, 2, 1)
plt.plot(months, sales, marker='o', color='blue', label='Sales')
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Monthly Sales")
plt.legend()
plt.grid(True)

# Second Subplot
plt.subplot(1, 2, 2)
plt.plot(months, profit, marker='s', color='green', label='Profit')
plt.xlabel("Months")
plt.ylabel("Profit")
plt.title("Monthly Profit")
plt.legend()
plt.grid(True)

# Adjust Layout
plt.tight_layout()

# Display Graph
plt.show()
```
