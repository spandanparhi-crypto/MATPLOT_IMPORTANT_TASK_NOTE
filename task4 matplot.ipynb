```python
import matplotlib.pyplot as plt

# Department salary data
departments = ['Engineering', 'Marketing', 'Sales', 'HR', 'Finance']

avg_salary = [95000, 75000, 70000, 65000, 85000]
max_salary = [150000, 110000, 120000, 90000, 130000]

# Create figure with 2 subplots
plt.figure(figsize=(12, 5))

# Plot 1: Average Salary
plt.subplot(1, 2, 1)
plt.bar(departments, avg_salary)
plt.title("Department Average Salary")
plt.xlabel("Departments")
plt.ylabel("Salary")
plt.xticks(rotation=20)

# Plot 2: Maximum Salary
plt.subplot(1, 2, 2)
plt.bar(departments, max_salary)
plt.title("Department Maximum Salary")
plt.xlabel("Departments")
plt.ylabel("Salary")
plt.xticks(rotation=20)

# Adjust layout
plt.tight_layout()

# Show plots
plt.show()
```
