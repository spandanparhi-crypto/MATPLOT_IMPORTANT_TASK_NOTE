```python
import pandas as pd

# Load Dataset
df = pd.read_csv("D:\Data science\Datasets\dataset.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))

sns.histplot(
    data=df,
    x='age',
    hue='target',
    kde=True,
    bins=15,
    palette=['skyblue', 'salmon'],
    alpha=0.6
)

plt.title('Age Distribution by Heart Disease Status')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Separate data by target
no_disease = df[df['target'] == 0]
disease = df[df['target'] == 1]

plt.figure(figsize=(10, 6))

# Scatter plot
plt.scatter(no_disease['chol'], no_disease['thalach'],
            color='blue', alpha=0.6, label='No Disease (0)')

plt.scatter(disease['chol'], disease['thalach'],
            color='red', alpha=0.6, label='Disease (1)')

# Regression line for No Disease
m1, b1 = np.polyfit(no_disease['chol'], no_disease['thalach'], 1)
plt.plot(no_disease['chol'],
         m1 * no_disease['chol'] + b1,
         color='blue')

# Regression line for Disease
m2, b2 = np.polyfit(disease['chol'], disease['thalach'], 1)
plt.plot(disease['chol'],
         m2 * disease['chol'] + b2,
         color='red')

plt.title('Cholesterol Levels vs Maximum Heart Rate')
plt.xlabel('Serum Cholesterol (chol)')
plt.ylabel('Maximum Heart Rate (thalach)')
plt.legend()
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Count chest pain types by target
cp_target = pd.crosstab(df['cp'], df['target'])

# Create bar plot
cp_target.plot(
    kind='bar',
    figsize=(8, 6),
    color=['skyblue', 'salmon']
)

plt.title('Chest Pain Type Distribution')
plt.xlabel('Chest Pain Type (cp)')
plt.ylabel('Number of Patients')
plt.legend(['No Disease (0)', 'Disease (1)'])
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Separate data by target
no_disease = df[df['target'] == 0]
disease = df[df['target'] == 1]

plt.figure(figsize=(10, 6))

# Scatter plot
plt.scatter(no_disease['age'], no_disease['trestbps'],
            color='blue', alpha=0.6, label='No Disease (0)')

plt.scatter(disease['age'], disease['trestbps'],
            color='red', alpha=0.6, label='Disease (1)')

# Trend line for No Disease
m1, b1 = np.polyfit(no_disease['age'], no_disease['trestbps'], 1)
plt.plot(no_disease['age'],
         m1 * no_disease['age'] + b1,
         color='blue')

# Trend line for Disease
m2, b2 = np.polyfit(disease['age'], disease['trestbps'], 1)
plt.plot(disease['age'],
         m2 * disease['age'] + b2,
         color='red')

plt.title('Resting Blood Pressure by Age and Target')
plt.xlabel('Age')
plt.ylabel('Resting Blood Pressure (trestbps)')
plt.legend()
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Prepare data for each group
female_no = df[(df['sex'] == 0) & (df['target'] == 0)]['thalach']
female_yes = df[(df['sex'] == 0) & (df['target'] == 1)]['thalach']
male_no = df[(df['sex'] == 1) & (df['target'] == 0)]['thalach']
male_yes = df[(df['sex'] == 1) & (df['target'] == 1)]['thalach']

# Create box plot
plt.figure(figsize=(8, 6))

plt.boxplot(
    [female_no, female_yes, male_no, male_yes],
    labels=[
        'Female\nNo Disease',
        'Female\nDisease',
        'Male\nNo Disease',
        'Male\nDisease'
    ],
    patch_artist=True
)

plt.title('Maximum Heart Rate by Sex and Target')
plt.xlabel('Sex and Target')
plt.ylabel('Maximum Heart Rate (thalach)')
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```
